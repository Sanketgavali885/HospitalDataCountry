```
df7=df_cleaned.select("HRR","Total Hospital Beds","Available Hospital Beds","Total ICU Beds","Available ICU Beds")
display(df7)
```

```
#cities and States are seperated from HRR column

df_split = df_with_custom_headers.withColumn("City", split(col("HRR"), ",")[0]) \
             .withColumn("State", split(col("HRR"), ",")[1])



df_split.display()
```

```
# Filtering statewise total and Available beds
df_split.createOrReplaceTempView("statewisedata")
df8=spark.sql("SELECT State, SUM(Total_Beds) AS Total_Beds,SUM(Available_Beds) AS Available_Beds, sum(Total_ICU)As Total_ICU_Beds,sum(Available_ICU)As Available_ICU_Beds FROM statewisedata GROUP BY State")
display(df8)
```
